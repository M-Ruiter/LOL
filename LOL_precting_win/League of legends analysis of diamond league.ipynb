# League of legends: what underlines a win

League of legends (LOL) is a strategic videogame in which two teams (red vs blue) of 5 players compete to destroy the opponents base. Each player chooses a champion with unique abilities. Killing the champions of the other team rewards gold, which can be used to buy item to enhance your champions strength. Champions respawn after a certain time. Typical in games, one team gets an advantage in kills - and subsequent gold income - that they become so stong that they overpower the other team and win the game.

The data set in this investigation collected from diamond and Master level. These are the most skilled players. You can compare this as the top league of a sport. As with sport, certain statistic can be extracted which can predict the win.

The statistic are gathered at the 10 minute mark. Games can last anywhere from 5 minutes up to 60 minutes or more. Average game lenght is roughly ~35 min. Furthermore, teams can surrender after 20 min if they feel that winning is no longer possible.



### Different columns in dataset
- gameId: Uniue game tag<br>
- blueWins.: Binary tag of blue winning<br>
- blueWardsPlaced: Number of wards placed (gives map vision)<br>
- blueWardsDestroyed: Ward removed by enemy team<br>
- blueFirstBlood: Binary tag of first kill. First kill give bonus gold<br>
- blueKills: Number of kills at 10 min<br>
- blueDeaths: Number of deaths at 10 min<br>
- blueAssists: Number of assists with kills. Helping another play land a kill give some gold. Less than a kill<br>
- blueEliteMonsters: Number of non-player neutral Monsters killed. Gives gold reward to a player <br>
- blueDragons: Dragon kills. Gives a strong buff and gold for whole team<br>
- blueHeralds: Special monster killed. Gives a strong buff and gold for whole team<br>     
- blueTowersDestroyed: Defensive structures. Needs to be destroyed in order to reach the base <br>
- blueTotalGold: Total gold
- blueAvgLevel: Level of champions. Champions get stronger during the game. The more kills, the faster this goes.
- blueTotalExperience: sum of all the experience champions aquired. Experience is needed for higher level
- blueTotalMinionsKilled: NPC that belong to both teams. They give gold when killed by players.
- blueTotalJungleMinionsKilled: neutral NPCs. They give gold when killed by players
- blueGoldDiff: Difference in gold with red team
- blueExperienceDiff: difference in experience with Red team
- blueCSPerMin: Minions per minute killed
- blueGoldPerMin: Total gold divided by game length

#### Same columns for red team, except:
- GameId
- RedWins: Only column BlueWins 


```python
# import modules and data.

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#%matplotlib inline

data = pd.read_csv('high_diamond_ranked_10min.csv')
data.head()
```

```python
data.info()
```

Data is clean without any missing values and all the columns have appropiate dtypes.

Several features seem clearly linked/correlated. For instance blueKills == redDeaths, so we have a lot of columns we can drop. Let's check the correlation between features.

```python
corr_matrix = data.drop(['gameId','blueWins'], axis=1).corr()
np.fill_diagonal(corr_matrix.values, np.nan)

s = corr_matrix.unstack()
so = s.sort_values(kind="quicksort", ascending=False)

high_corr = so[(so>0.9) | (so<-0.9)]
print(high_corr)
```

```python
# drop highly correlated columns
data.drop(['gameId', 'blueAvgLevel', 'blueCSPerMin', 'blueGoldPerMin',
          'redFirstBlood', 'redKills', 'redDeaths', 'redAvgLevel', 'redGoldDiff',
          'redExperienceDiff', 'redCSPerMin','redGoldPerMin'], axis=1, inplace=True)
```

```python
data.shape
```

```python
#Exploring remaining data

# Set up the matplotlib figure
plt.subplots(7, 4)
for i, col in enumerate(data.columns[1:]):
    plt.subplot(7, 4, i+1)
    sns.kdeplot(data.loc[data['blueWins'] == 0, col], shade=True, label='blue loses')
    sns.kdeplot(data.loc[data['blueWins'] == 1, col], shade=True, label='blue wins')
    plt.title(col)
    if i == 3:
        plt.legend(loc='upper right', fontsize ='x-large')
    else:
        plt.legend().remove()

plt.gcf().set_size_inches(20, 35)
```

## Conclusions from data exploration.

Most features are linked to winning. Clearest in <code>blueGoldDiff</code> and <code>blueExperienceDiff</code> 

Some features can be dropped as they have no clear effect on winning. These are:
- blueWardsPlaced
- blueWardsDestroyed
- redWardsPlaced
- redWardsDestroyed
- blueHeralds
- blueTowersDestroyed
- redHeralds
- redTowersDestroyed


```python
y = data.blueWins
X = data.drop(['blueWins','blueWardsPlaced', 'blueWardsDestroyed', 'redWardsPlaced', 'redWardsDestroyed',
          'blueHeralds', 'blueTowersDestroyed', 'redHeralds', 'redTowersDestroyed'], axis=1)
```

# Fitting some models

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import cross_val_score
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import AdaBoostClassifier


svc = SVC(gamma='auto')
LR = LogisticRegression(solver='lbfgs')
RFR = RandomForestClassifier(n_estimators=100)
GNB = GaussianNB()
Ada = AdaBoostClassifier(n_estimators=100)


models =[svc, LR, RFR, GNB, Ada]

accuracy={}

for model in models:
    cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
    K_scores = cross_val_score(model, X_scaled, y, cv=cv)
    accuracy[str(model)[:4]] = np.mean(K_scores)

accuracy = pd.Series(accuracy)
```

```python
print(accuracy)
```

```python
RFR.fit(X, y)

feature_importance = list(RFR.feature_importances_)
features={}

for n in range(len(feature_importance)):
    features[list(X)[n]] = feature_importance[n]*100
    
features = pd.Series(features)
features.sort_values(ascending=True).plot(kind='barh')
```

```python
from sklearn.ensemble import VotingClassifier

eclf1 = VotingClassifier(estimators=[('lr', LR), ('rf', RFR), ('gnb', GNB), ('svc', svc), ('Ada', Ada)], voting='hard')

cv = StratifiedShuffleSplit(n_splits=5, test_size=0.2, random_state=42)
K_scores = cross_val_score(eclf1, X, y, cv=cv)
print(np.mean(K_scores))
```
